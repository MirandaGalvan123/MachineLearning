```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# REcuerda estudiar proyectos data camp carpeta git hub

```python
houses = pd.read_csv('House.csv')
houses.head(2)
```

```python
sns.barplot(data=houses,x='Type',y='Price',estimator='mean')
#Por default el estimator es mean
#Por default agurpa todos los tipos a lo largo del tiempo
plt.show()
```

```python
sns.kdeplot(data=houses, x='Price',hue='Type',cut=0)
plt.show()
```

```python
sns.scatterplot(data=houses, x='Price',y='Lattitude',hue='Type')
plt.show()
```

```python
sns.histplot(data=houses, x='Price')
plt.show()
```
